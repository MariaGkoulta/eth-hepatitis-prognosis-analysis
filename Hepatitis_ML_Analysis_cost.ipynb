## Missing Value Imputation

## Scaling

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

#scaler = MinMaxScaler(feature_range=(0, 1))
scaler = StandardScaler()

X_df = pd.DataFrame(scaler.fit_transform(X_df.values), columns = X_df.columns)
```

```python
from sklearn.impute import SimpleImputer,  KNNImputer
import numpy as np

def impute_nan_values(X_df, imputation_strategy = "mean"):
  for column in X_df.columns:
    # unique_values = X_df[column].nunique()
    # if unique_values == 2: continue
    # else:
    if X_df[column].isnull().any():
        if imputation_strategy == "mean":
          imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
        elif imputation_strategy == "median":
          imputer = SimpleImputer(missing_values=np.nan, strategy='median')
        elif imputation_strategy == "knn":
          imputer = KNNImputer(missing_values=np.nan, n_neighbors=10)
        X_df[column] = imputer.fit_transform(X_df[[column]])
  return X_df
```

```python
X_df = impute_nan_values(X_df, imputation_strategy = "knn")
```

```python
from skopt.space import Real, Categorical, Integer
from skopt import BayesSearchCV
from sklearn.linear_model import Lasso, LassoCV, Ridge, RidgeCV, LogisticRegression


search_spaces = {'C': Real(0.005,0.1)}
model = LogisticRegression(penalty='l1',solver='liblinear')

opt = BayesSearchCV(estimator=model,
                    search_spaces=search_spaces,
                    scoring='accuracy',
                    cv=5,
                    n_iter=50,
                    n_jobs=1,
                    random_state=10,
                    verbose=10)


opt.fit(X,y)
```

# Expenses per patient

```python
# !! Assume here that the dataset is already imported, see first code snippet in this colab !!

X_orig = hepatitis.data.features
X_np = np.array(X_orig)
y_orig = hepatitis.data.targets
y_np = np.array(y_orig)
costs_all_patients = []

for i in range(0,len(y_np)):
  cost_per_patient = 0.0
  discount = False
  # more than one blood test was done -> discount
  if np.isnan(X_np[i,13:18]).sum() < 4:
    discount = True
    for j in range(0,len(X_np[i])):
      if discount and not np.isnan(X_np[i][j]):
        cost_per_patient+=costs['cost_group'][j]
  else:
    for j in range(0,len(X_np[i])):
      if not np.isnan(X_np[i][j]):
        cost_per_patient+=costs['cost_individual'][j]
  if discount:
    cost_per_patient+=2.10
  costs_all_patients.append(cost_per_patient)

costs_all_patients = np.round(np.array(costs_all_patients),decimals=2)
survived_costs = np.round(costs_all_patients[np.where(y_np == [2])[0]], decimals=2)
died_costs = np.round(costs_all_patients[np.where(y_np == [1])[0]], decimals=2)

print("Mean expenses for patients depending on outcome\n")
print("Patient survived: ", survived_costs.mean())
print("Patient died: ", died_costs.mean())
```

```python
"""
Idea 1: For each Tree, get the most important features and see if there is a correlation between more predictive feature and cost of the feature
"""

from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier

def pred_cost_correlation(model, modelname):
  model.fit(X, y)
  predictiveness = model.feature_importances_
  return predictiveness

def plot_pred_cost_corr(predictiveness):
  pred_per_feature = pd.DataFrame([[i] for i in predictiveness], columns = ['predictiveness'])
  cost_per_feature = pd.DataFrame([[i] for i in costs['cost_group']], columns = ['costs'])

  merged = pd.concat([pred_per_feature, cost_per_feature],axis=1)

  x = merged['predictiveness']
  y = merged['costs']

  """
  Ugly code copied from ChatGPT
  """

  # Calculate 2D histogram to get density of points at each position
  hist, x_edges, y_edges = np.histogram2d(x, y, bins=40)

  # Find the indices of the bin that each point belongs to
  x_indices = np.digitize(x, x_edges) - 1
  y_indices = np.digitize(y, y_edges) - 1

  # Count the number of points in each bin
  bin_counts = np.zeros_like(hist)
  for i in range(len(x)):
      # Ensure that indices are within the valid range
      x_idx = max(0, min(x_indices[i], len(bin_counts) - 1))
      y_idx = max(0, min(y_indices[i], len(bin_counts) - 1))
      bin_counts[x_idx, y_idx] += 1

  # Create an array to store colors for each point based on the density
  colors = np.zeros(len(x))
  for i in range(len(x)):
      # Ensure that indices are within the valid range
      x_idx = max(0, min(x_indices[i], len(bin_counts) - 1))
      y_idx = max(0, min(y_indices[i], len(bin_counts) - 1))
      colors[i] = bin_counts[x_idx, y_idx]

  # Plot the scatter plot with colored points
  plt.scatter(x, y, c=colors, cmap='viridis')
  plt.colorbar(label='Density')





  plt.xlabel('Predictiveness')
  plt.ylabel('Cost')
  plt.title('Correlation between Predictiveness and Cost')
  plt.grid(True)
  plt.show()

  cor_matrix = merged.corr()
  plt.figure(figsize =(15, 7))
  dataplot=sns.heatmap(cor_matrix,annot=True, fmt=".1f")
  plt.show()
```
